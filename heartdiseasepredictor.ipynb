```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import neighbors
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("heart.csv")
df.info()
```

```python
y=df["target"]
x=df.drop("target",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
#svc model with rbf kernel curve
model1=SVC(kernel="rbf")
model1.fit(x_train,y_train)
test_out1=model1.predict(x_test)
acc1=accuracy_score(test_out1,y_test)
print(acc1)
```

```python
#svc model(linear kernel)
model2=SVC(kernel="linear")
model2.fit(x_train,y_train)
model2_out=model2.predict(x_test)
acc2=accuracy_score(model2_out,y_test)
print(acc2)
```

```python
#svc model with poly kernel
model3=SVC(kernel="poly")
model3.fit(x_train,y_train)
model3_out=model3.predict(x_test)
acc3=accuracy_score(y_test,model3_out)
print(acc3)
```

```python
#svc model with sigmoid kernel
model4=SVC(kernel="poly")
model4.fit(x_train,y_train)
model4_out=model4.predict(x_test)
acc4=accuracy_score(y_test,model4_out)
print(acc4)
```

```python
#linear classifier
model5=SGDClassifier()
model5.fit(x_train,y_train)
model5_out=model5.predict(x_test)
acc5=accuracy_score(y_test,model5_out)
print(acc5)
```

```python
#knn model
model6=neighbors.KNeighborsClassifier(n_neighbors=21)
model6.fit(x_train,y_train)
model6_out=model6.predict(x_test)
acc6=accuracy_score(y_test,model6_out)
print(acc6)
```

```python
#decision tree model
model7=DecisionTreeClassifier()
model7.fit(x_train,y_train)
model7_out=model7.predict(x_test)
acc7=accuracy_score(y_test,model7_out)
print(acc7)
```

```python
#radom forest classifier
model8=RandomForestClassifier(max_depth=2, random_state=0)
model8.fit(x_train,y_train)
model8_out=model8.predict(x_test)
acc8=accuracy_score(y_test,model7_out)
print(acc8)
```

```python
x=["svc\nrbf","svc\nlinear","svc\npoly","svc\nsigmoid","linear\nclassifier","knn","decision\ntree","random\nforest\nclassifier"]
yaxis=[acc1,acc2,acc3,acc4,acc5,acc6,acc7,acc8]
plt.xlabel("Machine learning algorithm",fontsize=12)
plt.ylabel("Accuracy",fontsize=12)
plt.xticks(rotation=20)
plt.bar(x,yaxis,width=0.4)
```
